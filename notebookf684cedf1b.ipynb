## 2. Exploratory Data Analysis

```python
# import all of the python modules/packages you'll need here
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

from collections import Counter
import string
from textblob import TextBlob
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
```

By displaying the dataset that I am working with and seeing the size of the products I can better visualize the data set.  

```python
dataset_path = '/kaggle/input/amazon-books-reviews/Books_rating.csv'
df = pd.read_csv(dataset_path)
print(f"df.shape: {df.shape}")
pd.set_option("display.max_colwidth", 240)
df.head()
```

First, I want to know how many reviews there are for each individual book 

```python
df['Title'].value_counts()
```

Since there are several books with only one review and we will not be able to gather much information from that I will remove any reviews with less than 20 reviews and recheck the size of the data set after those are removed. 

```python
df = df[df.groupby("Title")['Title'].transform("size") > 20]
len(df)
```

Since the data set is still rather large I will only focus on the first 10,000 entries

```python
df = df.head(10000)
```

**Preprocessing**

I will convert the review/text column to a string if not in a string format already.  I will also remove any row that is missing a review

```python
df['review/text'] = df['review/text'].astype(str)
df = df[~df['review/text'].isnull()]
len(df)
```

This section will cleans the data set and replace all special characters with apostrophes.

```python
def clean(txt):
    txt = txt.str.replace(r"\(", "")
    txt = txt.str.replace(r'(<a).*(>).*()', '')
    txt = txt.str.replace(r'(&amp)', '')
    txt = txt.str.replace(r'(&gt)', '')
    txt = txt.str.replace(r'(&lt)', '')
    txt = txt.str.replace(r'(\xa0)', ' ')  
    return txt
df['review/text'] = clean(df['review/text'])
```

This section will convert all reviews to lower case to ensure words are only counted once.  

```python
df['review1'] = df['review/text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df['review1'].head()
```

This section will remove all punctuation to also help in counting words only once 

```python
import re

def remove_punctuation(text):
    return re.sub(r'[^\w\s]', '', text)

df['review1'] = df['review1'].apply(remove_punctuation)
df['review1'].head()
```

Here we are removing all stopwords from reviews

```python

nltk.download('stopwords')

stop = stopwords.words('english')
df['review1'] = df['review1'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df['review1'].head()
```

In this section we will first display any unique words(words that only appear once in a review) and then remove these unique words.

```python
freq = pd.Series(' '.join(df['review1']).split()).value_counts()
less_freq = list(freq[freq ==1].index)
first_25_words = less_freq[:25]
print(first_25_words)
```

```python
df['review1'] = df['review1'].apply(lambda x: " ".join(x for x in x.split() if x not in less_freq))
```

**Data Analysis**

Length of review

```python
df['review_len'] = df['review/text'].astype(str).apply(len)
df['word_count'] = df['review/text'].apply(lambda x: len(str(x).split()))
```

Show the sentiment(positive, neutral, negative) of a review

```python
df['polarity'] = df['review1'].map(lambda text: TextBlob(text).sentiment.polarity)
df.head()
```

Displays histograms that show the review length, word count, and the sentiment(polarity) of a review.

```python
df[["review_len", "word_count", "polarity"]].hist(bins=20, figsize=(15, 10))
```

Box plot showing Polarity vs Rating on Amazon to see if there are any differences between the review scores and the data analysis performed above. 

```python

plt.figure(figsize = (10, 8))
sns.set_style('whitegrid')
sns.set(font_scale = 1.5)
sns.boxplot(x = 'review/score', y = 'polarity', data = df)
plt.xlabel("Rating")
plt.ylabel("Polatiry")
plt.title("Product Ratings vs Polarity")
plt.show()
```

```python
mean_pol = df.groupby('review/score')['polarity'].agg([np.mean])
mean_pol.columns = ['mean_polarity']
fig, ax = plt.subplots(figsize=(8, 6))
plt.bar(mean_pol.index, mean_pol.mean_polarity, width=0.3)
#plt.gca().set_xticklabels(mean_pol.index, fontdict={'size': 14})
for i in ax.patches:
    ax.text(i.get_x(), i.get_height()+0.01, str("{:.2f}".format(i.get_height())))
```

Top most occurring words in the reviews

```python

def get_top_n_words(corpus, n=None):
    vec=CountVectorizer().fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]

common_words = get_top_n_words(df['review1'], 20)
df1 = pd.DataFrame(common_words, columns = ['Review', 'count'])
df1.head()
```
